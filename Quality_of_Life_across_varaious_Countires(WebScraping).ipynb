```python
#Importing all the required libraries which will be Helpful in Project
import numpy as np  # Used to perform Advance mathematical functions

import pandas as pd #data manipulation and exploration tool

import matplotlib.pyplot as plt #data visulization

import seaborn as sns           #data visulization

import requests                 #to access the url and check the status

from bs4 import BeautifulSoup   #Used to scrap the data

import re                       #Mojorly used to clean the data and to extract the useful and meaninful data from raw data

```

```python
#Checking the status code i.e wheater content of website is accessible or not

url='https://www.numbeo.com/quality-of-life/rankings_by_country.jsp'
req=requests.get(url)
req.status_code
```

```python
#Scraping or extracting the raw data from webstite to Soup Variable with Help of Beautiful Soup Library
soup=BeautifulSoup(req.content)
soup
```

```python
''' Trying to Extract Name of the Countries ,
Quality of life Index , Purchasing power Index , Helath care Index,
Safety Index , Cost of Living , Pollution Index and Climate Index  and required data From the raw data with The help of regex library'''



raw_array=[]
for element in soup.find_all('td'):
      raw_array.append(element)
raw_array

```

```python
#Extracting only the required data from whole of data
pattern='^\[.*?<td></td>,|<span class="pre_language.*?\]|<\/td>|<td.*?>'
raw_array=re.sub(pattern, '',str(raw_array), flags=re.DOTALL)
raw_array

#Make the list of extracted data and seprating each element
Actual_array=raw_array.split(',')

for i in Actual_array:
  if i=='' or i==' ':
    Actual_array.remove(i)

Actual_array
```

```python
# Creating the Columns for particular type of data
Countries=[]
Quality_of_life_Index=[]
Purchasing_power_Index=[]
Safety_Index=[]
Helath_care_Index=[]
Cost_of_Living=[]
Property_Income_Ratio=[]
Traffic_Commute_Time_Index=[]
Pollution_Index=[]
Climate_Index=[]
```

```python
Actual_array
```

```python
#Appending and adding the data in newly created columns
count=0
for i in range(0,len(Actual_array)):
  if count==0:
    Countries.append(Actual_array[i])
    count+=1;

  elif count==1:
    Quality_of_life_Index.append(Actual_array[i])
    count+=1;

  elif count==2:
    Purchasing_power_Index.append(Actual_array[i])
    count+=1

  elif count==3:
    Safety_Index.append(Actual_array[i])
    count+=1

  elif count==4:
    Helath_care_Index.append(Actual_array[i])
    count+=1

  elif count==5:
    Cost_of_Living.append(Actual_array[i])
    count+=1

  elif count==6:
    Property_Income_Ratio.append(Actual_array[i])
    count+=1

  elif count==7:
    Traffic_Commute_Time_Index.append(Actual_array[i])
    count+=1

  elif count==8:
    Pollution_Index.append(Actual_array[i])
    count+=1

  elif count==9:
    Climate_Index.append(Actual_array[i])
    count=0


```
